```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('student_data_large.csv')
df
```

```python
# perfrom different operation on the dataframe
# Display basic information about the dataframe 

df = df.rename(columns={"Age":"Umar"})
df
```

```python
df.boxplot(column='Height', by='Umar')
plt.title('Height Distribution by Umar (Age)')
plt.show()
```

```python
plt.hist(df['Height'], bins=df['Umar'].nunique())
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.title('Height Distribution')
plt.show()
```

```python
plt.pie(df["Height"].value_counts().head(5), labels=df["Height"].value_counts().head(5).index, autopct='%1.1f%%')
```

```python
plt.barplot(df["Height","Course"])
```
