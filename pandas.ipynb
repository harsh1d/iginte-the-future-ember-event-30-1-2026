```python
# Data Visualization in Python using Pandas and Matplotlib

import matplotlib.pyplot as plt
import seaborn as sns
```

```python

plt.figure(figsize=(6,5))
sns.countplot(x="Sex", hue="Survived", data=read_csv, palette={0:"#fc8d62", 1:"#66c2a5"})
plt.title("Survival by gender")
plt.xlabel("Gender")
plt.legend(["Did not survived ","Survived"])
plt.ylabel("Count")
plt.show()
```
