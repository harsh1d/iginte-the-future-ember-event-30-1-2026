```python
def calculator(a, b, operation):
    if operation == '+':
        return a + b
    elif operation == '-':
        return a - b
    elif operation == '*':
        return a * b
    elif operation == '/':
        return a / b if b != 0 else "Error: Division by zero"
    else:
        return "Error: Invalid operation"

# Test the calculator
print(calculator(10, 5, '+'))
print(calculator(10, 5, '-'))
print(calculator(10, 5, '*'))
print(calculator(10, 5, '/'))
```

```python
# Dictionary operations with examples

# Complete the student dictionary with more key-value pairs
student = {
	"name": "harsh",
	"age": 20,
	"course": "Python Programming",
	"grade": "A"
}

print("Original Dictionary:", student)

# Accessing values
print("\nAccessing values:")
print("Name:", student["name"])
print("Age:", student.get("age"))

# Adding new key-value pairs
student["email"] = "harsh@example.com"
print("\nAfter adding email:", student)

# Updating values
student["age"] = 21
print("\nAfter updating age:", student)

# Removing items
student.pop("grade")
print("\nAfter removing grade:", student)

# Dictionary methods
print("\nDictionary keys:", student.keys())
print("Dictionary values:", student.values())
print("Dictionary items:", student.items())

# Checking if key exists
print("\nIs 'name' in dictionary?", "name" in student)
print("Is 'grade' in dictionary?", "grade" in student)

# Dictionary comprehension example
squares = {x: x**2 for x in range(1, 6)}
print("\nSquares dictionary:", squares)

# Nested dictionary
students = {
	"student1": {"name": "harsh", "age": 20},
	"student2": {"name": "prashant", "age": 22}
}
print("\nNested dictionary:", students)
print("Student1 name:", students["student1"]["name"])
```

```python
# Dictionary methods with examples

# Create a sample dictionary
person = {
    "name": "Alice",
    "age": 25,
    "city": "New York",
    "profession": "Engineer"
}

print("Original Dictionary:", person)
print("\n" + "="*50 + "\n")

# 1. keys() - Returns all keys
print("1. keys() method:")
print("All keys:", person.keys())
print()

# 2. values() - Returns all values
print("2. values() method:")
print("All values:", person.values())
print()

# 3. items() - Returns key-value pairs
print("3. items() method:")
print("All items:", person.items())
print()

# 4. get() - Returns value for a key (safe access)
print("4. get() method:")
print("Age:", person.get("age"))
print("Salary (doesn't exist):", person.get("salary", "Not Found"))
print()

# 5. update() - Updates dictionary with another dictionary
print("5. update() method:")
person.update({"age": 26, "country": "USA"})
print("After update:", person)
print()

# 6. pop() - Removes and returns value for a key
print("6. pop() method:")
removed_value = person.pop("profession")
print("Removed value:", removed_value)
print("After pop:", person)
print()

# 7. popitem() - Removes and returns last key-value pair
print("7. popitem() method:")
last_item = person.popitem()
print("Last item removed:", last_item)
print("After popitem:", person)
print()

# 8. clear() - Removes all items
print("8. clear() method:")
temp_dict = {"a": 1, "b": 2}
print("Before clear:", temp_dict)
temp_dict.clear()
print("After clear:", temp_dict)
print()

# 9. copy() - Creates a shallow copy
print("9. copy() method:")
person_copy = person.copy()
print("Original:", person)
print("Copy:", person_copy)
print()

# 10. setdefault() - Returns value if key exists, else sets and returns default
print("10. setdefault() method:")
person.setdefault("age", 30)
person.setdefault("email", "alice@example.com")
print("After setdefault:", person)
print()

# 11. fromkeys() - Creates dictionary from keys with same value
print("11. fromkeys() method:")
keys = ["x", "y", "z"]
new_dict = dict.fromkeys(keys, 0)
print("Dictionary from keys:", new_dict)
```

```python
# Dictionary methods demonstrated with student1
student1 = {"maths": 100}

print("Original Dictionary:", student1)
print("\n" + "="*50 + "\n")

# 1. keys() - Returns all keys
print("1. keys() method:")
print("All keys:", student1.keys())
print()

# 2. values() - Returns all values
print("2. values() method:")
print("All values:", student1.values())
print()

# 3. items() - Returns key-value pairs
print("3. items() method:")
print("All items:", student1.items())
print()

# 4. get() - Returns value for a key (safe access)
print("4. get() method:")
print("Maths marks:", student1.get("maths"))
print("Science marks (doesn't exist):", student1.get("science", "Not Found"))
print()

# 5. update() - Updates dictionary with another dictionary
print("5. update() method:")
student1.update({"science": 95, "english": 88})
print("After update:", student1)
print()

# 6. pop() - Removes and returns value for a key
print("6. pop() method:")
english_marks = student1.pop("english")
print("Removed english marks:", english_marks)
print("After pop:", student1)
print()

# 7. popitem() - Removes and returns last key-value pair
print("7. popitem() method:")
last_subject = student1.popitem()
print("Last item removed:", last_subject)
print("After popitem:", student1)
print()

# 8. setdefault() - Returns value if key exists, else sets and returns default
print("8. setdefault() method:")
student1.setdefault("maths", 50)
student1.setdefault("physics", 92)
print("After setdefault:", student1)
print()

# 9. copy() - Creates a shallow copy
print("9. copy() method:")
student1_copy = student1.copy()
print("Original:", student1)
print("Copy:", student1_copy)
print()

# 10. clear() - Removes all items (demonstrated on copy)
print("10. clear() method:")
print("Before clear:", student1_copy)
student1_copy.clear()
print("After clear:", student1_copy)
print()

# 11. fromkeys() - Creates dictionary from keys with same value
print("11. fromkeys() method:")
subjects = ["maths", "science", "english", "history"]
default_marks = dict.fromkeys(subjects, 0)
print("Dictionary from keys:", default_marks)
print()

# 12. in operator - Check if key exists
print("12. Check if key exists:")
print("Is 'maths' in student1?", "maths" in student1)
print("Is 'history' in student1?", "history" in student1)
print()

# 13. len() - Get number of key-value pairs
print("13. len() function:")
print("Number of subjects:", len(student1))
```
